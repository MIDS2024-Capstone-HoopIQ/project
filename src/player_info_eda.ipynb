# Look at Birth Year and College Distributions

```python
athlete_df[athlete_df['athlete_birthdate'].isna()]
```

```python
birthyear_df = athlete_df[['athlete_birthdate']]
birthyear_df = birthyear_df.dropna(subset=['athlete_birthdate'])
birthyear_df['birthyear'] = pd.to_datetime(birthyear_df['athlete_birthdate']).dt.year
birthyear_df = birthyear_df.drop(columns=['athlete_birthdate'])
```

```python
year_counts = birthyear_df['birthyear'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar')
plt.title('Year Frequency')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
athlete_df[athlete_df['athlete_college'].isna()]
```

```python
college_df = athlete_df[['athlete_college']]
college_df = college_df.dropna(subset=['athlete_college'])
```

```python
college_counts = college_df['athlete_college'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(20, 6))
college_counts.plot(kind='bar')
plt.title('College Frequency')
plt.xlabel('College')
plt.ylabel('Count')
plt.show()
```
